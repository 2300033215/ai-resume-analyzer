```python
import os
from PyPDF2 import PdfReader
from pydantic_ai import Agent
from pydantic_ai.settings import ModelSettings
from google.colab import files
import nest_asyncio

# Enable async for Colab
nest_asyncio.apply()

# Upload new resume file
uploaded = files.upload()
resume_file = list(uploaded.keys())[0]
print(f"📄 Uploaded: {resume_file}")

# Extract text from PDF
def extract_text_from_pdf(file_path):
    try:
        reader = PdfReader(file_path)
        text = ""
        for page in reader.pages:
            text += page.extract_text() + "\n"
        return text
    except Exception as e:
        return f"❌ Error reading PDF: {e}"

resume_text = extract_text_from_pdf(resume_file)

# Show preview
print("\n📑 Resume Preview:\n")
print(resume_text[:500])

# Set API key (if not already set)
if "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" not in os.environ:
    os.environ["GROQ_API_KEY"] = "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"  # Replace this

# Define AI Agent
resume_agent = Agent(
    model="groq:llama3-8b-8192",
    model_settings=ModelSettings(temperature=0.2),
    output_type=str,
    system_prompt="""
You are an AI resume reviewer. Given the resume's raw text, provide:
1. Missing key skills (technical & soft)
2. Formatting issues (if any)
3. Job-readiness score out of 10
4. Suggestions to improve the resume
Be clear and professional.
"""
)

# Analyze resume
response = resume_agent.run_sync(resume_text)

# Show feedback
print("\n🧠 AI Resume Feedback:\n")
print(response.output)
```
